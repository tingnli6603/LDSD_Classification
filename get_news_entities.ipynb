### 利用DBpedia Spotlight取得所有文本的Entity

先將會用到的文本放入DBpedia Spotlight取得Entity，之後就利用Entity代表這篇文章的內容。

>要注意編碼問題，如果沒有response.encoding = 'utf-8'會出現亂碼，導致後面找value時出錯。

```python
import os
import requests
from bs4 import BeautifulSoup

#開啟文本檔案
class_list = ['business','entertainment','politics','sport','tech']

for class_name in class_list:
    
    print(class_name)

    file_path = "./bbc/"+class_name

    class_name_list = os.listdir(file_path)
    
    for file_name in class_name_list:
            
        print(file_name)

        news_file_path = os.path.join(file_path,file_name)
        news_content_open = open(news_file_path,"r",encoding="iso-8859-15")
        
        news_content_line = news_content_open.readline()
        
        text_uri_list = []
        
        while news_content_line:
        
            #設定參數
            #自行架設DBpedia Spotlight服務
            #將文本檔案放入服務當中
            url_word = "http://XXX.XXX.XXX.XXX/rest/candidates?text="
            confnsup_settings = "&confidence=0.5&support=20"
            response = requests.get(url_word+news_content_line+confnsup_settings)
            response.encoding = 'utf-8'
            response_text = response.text

            #利用BeautifulSoup解析輸出，取得我們所需要的Entity數值
            soup = BeautifulSoup(response_text)
            part_uri_list = []
            for resource in soup.find_all('resource'):

                uri_text = resource['uri']
                part_uri_list.append(uri_text)
                
            text_uri_list = text_uri_list + part_uri_list
                
            news_content_line = news_content_open.readline()


        text_uri_dict = {}
        for entity in text_uri_list:
            
            entity = entity.encode('utf-8')
            if entity in text_uri_dict:
                text_uri_dict[entity] +=1
            else:
                text_uri_dict[entity] = 1
                
        #print(text_uri_dict)
            
        f = open("./all_entities/"+class_name+"/"+file_name, "w")
        f.write(str(text_uri_dict))
        f.close()
```
