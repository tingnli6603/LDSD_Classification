```python
import os
```

```python
topic_name_list = ['business','entertainment','politics','sport','tech']
```

### 每一個類別都是二元分類，依照每個類別切割解答

```python
test_entities_path = "../test_entities"
test_entities_name_list = os.listdir(test_entities_path)

for i in range(5):
    topic_name = topic_name_list[i]
    print(topic_name)
    
    topic_answer_dict = {}
    for file_name in test_entities_name_list:
        if topic_name in file_name:
            topic_answer_dict[file_name] = 1
        else:
            topic_answer_dict[file_name] = 0
    
    f = open(topic_name+"_answer.txt","w")
    f.write(str(topic_answer_dict))
    f.close()
```

### 預測檔案也依照類別切割

```python
predict_result_open = open("./classification_result_score.txt","r")
predict_result_dict = eval(predict_result_open.read())

for i in range(5):
    topic_name = topic_name_list[i]
    #print(topic_name)
    
    predict_dict = {}
    for file_name in predict_result_dict:
        print(file_name)

        predict_score_list = predict_result_dict[file_name]
        print(predict_score_list)
        min_score = min(predict_score_list)
        print(min_score)

        #找到該主題在該news的score
        score = predict_score_list[i]
        print(score)
        
        #是否該score就是最小值
        if score == min_score:
            predict_dict[file_name] = 1
        else:
            predict_dict[file_name] = 0

    g = open(topic_name+"_predict.txt","w")
    g.write(str(predict_dict))
    g.close()
```
